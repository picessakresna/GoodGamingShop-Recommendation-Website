## Data Cleaning & Pre Processing:

```python
df = df.rename(columns={
    'Nama Produk': 'nama_produk',
    'Review': 'review',
    'Nama Akun': 'nama_akun',
    'Rating User': 'rating_user',
    'Ulasan Produk': 'ulasan_produk',
    'Waktu Review': 'waktu_review',
    'Varian Produk': 'varian_produk'
})

df.columns
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
```

```python
df_produk = pd.read_csv('../data-collection-preprocessing/data-produk/clean_product-goodgamingshop.csv')
df_produk.head()
```

```python
df = df.merge(df_produk[['id_produk', 'nama_produk']], on='nama_produk', how='left')
df.head()
```

```python
kolom_terurut = ['id_produk'] + [col for col in df.columns if col != 'id_produk']
df = df[kolom_terurut]
df.head()
```

```python
df.columns
```

```python
# Dapatkan nama akun unik
unique_accounts = df['nama_akun'].unique()

# Buat mapping antara nama akun dan id user menggunakan hash
user_id_mapping = {nama_akun: idx + 1 for idx, nama_akun in enumerate(unique_accounts)}

# Tambahkan kolom id_user ke DataFrame utama
df.insert(3, 'id_user', df['nama_akun'].map(user_id_mapping))

df.head()
```

```python
def transform_id_user(id):
    return f'UI{id:04d}'

df['id_user'] = df['id_user'].apply(transform_id_user)
```

```python
df[df['id_user'] == 'UI0088']
```

```python
# Langkah 1: Temukan nama akun yang sama
duplicate_accounts = df[df.duplicated(subset=['id_user'], keep=False)]

# Langkah 2: Kelompokkan berdasarkan nama akun dan filter grup yang memiliki lebih dari satu produk berbeda
diff_products = duplicate_accounts.groupby('id_user').filter(lambda x: x['id_produk'].nunique() > 1)

diff_products
```

```python
rating_mapping = {
    'bintang 5': 5,
    'bintang 4': 4,
    'bintang 3': 3,
    'bintang 2': 2,
    'bintang 1': 1
}

df['rating_user'] = df['rating_user'].replace(rating_mapping).astype('int64')
```

```python
df.to_csv('../data-collection-preprocessing/data-ulasan-clean/clean_data-ulasan-goodgamingstore.csv', index=False)
```
